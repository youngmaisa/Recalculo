```python
import os
import pandas as pd

# Ruta de la carpeta donde están los archivos CSV
ruta_registros = r"C:\Users\mmedinam\OneDrive - Entel Peru S.A\Documentos - EntelDrive_ __ Canal Masivo\Archivos_Recalculo\ACTIVIDAD_SS_MASIVO\2024\Anio"

# Ruta de salida para los archivos Excel
ruta_salida_base = r"C:\Users\mmedinam\OneDrive - Entel Peru S.A\Documentos - EntelDrive_ __ Canal Masivo\Archivos_Recalculo\ACTIVIDAD_SS_MASIVO\2024\Meses"

# Diccionario para convertir el número de mes a nombre
meses = {
    '01': 'Enero', '02': 'Febrero', '03': 'Marzo', '04': 'Abril',
    '05': 'Mayo', '06': 'Junio', '07': 'Julio', '08': 'Agosto',
    '09': 'Septiembre', '10': 'Octubre', '11': 'Noviembre', '12': 'Diciembre'
}

# Paso 1: Listar los archivos CSV en la carpeta
archivos = [f for f in os.listdir(ruta_registros) if f.startswith('Registros_') and f.endswith('.csv')]

# Paso 2: Procesar cada archivo encontrado
for archivo in archivos:
    # Extraer el año del nombre del archivo (suponiendo formato registros_YYYY.csv)
    año = archivo.split('_')[1].split('.')[0]  # Esto obtiene el año, como '2024', '2023', etc.
    
    # Leer el archivo CSV
    archivo_path = os.path.join(ruta_registros, archivo)
    try:
        df = pd.read_csv(archivo_path)
        print(f"Archivo {archivo} leído correctamente.")
    except Exception as e:
        print(f"No se pudo leer el archivo {archivo}: {e}")
        continue

    # Paso 3: Crear la carpeta para el año si no existe
    ruta_salida_año = os.path.join(ruta_salida_base, año)
    if not os.path.exists(ruta_salida_año):
        os.makedirs(ruta_salida_año)

    # Paso 4: Filtrar por mes y guardar en archivos Excel
    for mes in sorted(df['NUMPERIODO'].astype(str).str[4:6].unique()):
        # Filtrar los registros correspondientes a ese mes
        df_mes = df[df['NUMPERIODO'].astype(str).str[4:6] == mes]
        
        if not df_mes.empty:  # Solo guardar si hay datos
            # Obtener el nombre del mes
            nombre_mes = meses.get(mes, f'Mes_{mes}')
            
            # Crear el archivo Excel
            archivo_mes = os.path.join(ruta_salida_año, f"Registros_{nombre_mes}.xlsx")
            
            # Guardar el DataFrame en un archivo Excel
            df_mes.to_excel(archivo_mes, index=False)
            print(f"Archivo guardado: {archivo_mes}")

print("Proceso completado.")

```

```python
import os
import pandas as pd

# Ruta de los archivos CSV
ruta_archivos_csv = r"C:\Users\mmedinam\OneDrive - Entel Peru S.A\Documentos - EntelDrive_ __ Canal Masivo\Archivos_Recalculo\ACTIVIDAD_SS_MASIVO\Registros_Ordenados\2024\Registros"
# Ruta donde se guardarán los nuevos archivos Excel
ruta_salida = r"C:\Users\mmedinam\OneDrive - Entel Peru S.A\Documentos - EntelDrive_ __ Canal Masivo\Archivos_Recalculo\ACTIVIDAD_SS_MASIVO\Registros_Ordenados\2024\Registro_Cruce"

# Crear la carpeta de salida si no existe
os.makedirs(ruta_salida, exist_ok=True)

# Listar los archivos .csv en la carpeta
archivos_csv = [f for f in os.listdir(ruta_archivos_csv) if f.endswith('.csv')]

for archivo in archivos_csv:
    archivo_path = os.path.join(ruta_archivos_csv, archivo)
    
    try:
        # Leer el archivo CSV
        df = pd.read_csv(archivo_path, encoding='utf-8', on_bad_lines='skip').dropna(how='all')
        
        # Asegurarnos de que las columnas necesarias existen
        columnas_necesarias = ['VCHDOCUMENTO', 'PERMANENCIA_2', 'FLAG_SUS_M2']
        if not all(col in df.columns for col in columnas_necesarias):
            print(f"Archivo {archivo} no contiene todas las columnas necesarias.")
            continue
        
        # Agrupar por 'VCHDOCUMENTO' y calcular métricas
        grupo = df.groupby('VCHDOCUMENTO').agg(
            SS=('VCHDOCUMENTO', 'size'),  # Número de veces que aparece cada VCHDOCUMENTO
            PERMANENCIA_2_SUM=('PERMANENCIA_2', 'sum'),
            FLAG_SUS_M2_SUM=('FLAG_SUS_M2', 'sum')
        ).reset_index()
        
        # Calcular la columna 'SUS M2 %'
        grupo['SUS M2 %'] = (grupo['FLAG_SUS_M2_SUM'] / grupo['SS']).round(2)
        
        # Nombre del archivo de salida
        nombre_salida = os.path.splitext(archivo)[0] + "_Cruce.xlsx"
        archivo_salida_path = os.path.join(ruta_salida, nombre_salida)
        
        # Guardar el resultado en un archivo Excel
        with pd.ExcelWriter(archivo_salida_path, engine='openpyxl') as writer:
            grupo.to_excel(writer, index=False, sheet_name='tabla_para_cruce')
        
        print(f"Archivo procesado y guardado: {archivo_salida_path}")
    
    except Exception as e:
        print(f"Error procesando el archivo {archivo}: {e}")

print("Proceso completado.")
```

```python
import pandas as pd
import os

# Ruta del archivo
ruta_archivo = r"C:\Users\mmedinam\OneDrive - Entel Peru S.A\Documentos - EntelDrive_ __ Canal Masivo\Archivos_Recalculo\ACTIVIDAD_SS_MASIVO\Actividad_SS_Masivo.csv"

# Leer el archivo CSV
try:
    df = pd.read_csv(ruta_archivo)
    print("Archivo leído correctamente.")
except Exception as e:
    print(f"Error al leer el archivo: {e}")

# Paso 1: Filtrar los registros del año 2024
df['NUMPERIODO'] = df['NUMPERIODO'].astype(str)  # Asegurarse de que la columna sea de tipo cadena
df_2024 = df[df['NUMPERIODO'].str.startswith('2024')]

# Contar los registros totales para 2024
total_registros_2024 = len(df_2024)

# Paso 2: Contar los registros por mes
df_2024['Mes'] = df_2024['NUMPERIODO'].str[4:6]  # Extraer el mes
registros_por_mes = df_2024['Mes'].value_counts().sort_index()  # Contar registros por mes y ordenar

# Mostrar resultados
print("\n###########################")
print(f"Total de registros para 2024: {total_registros_2024}")
print("Registros por mes:")
print(registros_por_mes)

```

```python
df_mes = df_2024[df_2024['Mes'] == '08']


vchdocumento_counts = df_mes['VCHDOCUMENTO'].value_counts().reset_index()
vchdocumento_counts.columns = ['VCHDOCUMENTO', 'SS']


df_agrupado = df_mes.groupby('VCHDOCUMENTO').agg({
    'FLAG_SUS_M2': 'sum',  
    'PERMANENCIA_2': 'sum'  
}).reset_index()


df_agrupado = pd.merge(vchdocumento_counts, df_agrupado, on='VCHDOCUMENTO')


df_agrupado['SUS M2 %'] = ((df_agrupado['FLAG_SUS_M2'] + df_agrupado['PERMANENCIA_2']) / df_agrupado['SS']).round(2)


print("\nResultados por 'VCHDOCUMENTO' (DNI) con el cálculo de 'SUS M2 %':")
print(df_agrupado[['VCHDOCUMENTO', 'SS', 'FLAG_SUS_M2', 'PERMANENCIA_2', 'SUS M2 %']])
```
