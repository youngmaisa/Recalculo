```python
import pandas as pd
import numpy as np

import streamlit as st
```

```python
df = pd.read_excel("ANÁLISIS CALIDAD PROACTIVO - CANALES.xlsx", sheet_name="Hoja")
df
```

```python
def asignar_deciles_y_rangos(dataframe):
   
    """
    deciles = pd.qcut(grupo['QVENTAS'], q=10, duplicates='drop')
    grupo['RANGO'] = deciles.apply(lambda x: f"({int(x.left)}-{int(x.right)}]")
    grupo['DECIL'] = pd.qcut(grupo['QVENTAS'], q=10, duplicates='drop', labels=range(1, len(deciles.cat.categories) + 1))
    grupo = grupo.sort_values('DECIL')
    return grupo
    """

    dataframe = dataframe.copy()
    deciles = pd.qcut(dataframe['QVENTAS'], q=10, duplicates='drop')
    dataframe['RANGO'] = deciles.apply(lambda x: f"({int(x.left)}-{int(x.right)}]")
    dataframe['DECIL'] = pd.qcut(dataframe['QVENTAS'], q=10, duplicates='drop', labels=False) + 1
    return dataframe


#ventas = ventas.groupby('TIPO').apply(asignar_deciles_con_rangos)

# Sidebar de filtros
st.sidebar.title("Filtros")
tipos = df['TIPO'].unique()
tipo_seleccionado = st.sidebar.selectbox("Selecciona el TIPO", tipos)

rango_min = st.sidebar.number_input("QVENTAS mínimo", min_value=int(df['QVENTAS'].min()), value=int(df['QVENTAS'].min()))
rango_max = st.sidebar.number_input("QVENTAS máximo", min_value=int(df['QVENTAS'].min()), value=int(df['QVENTAS'].max()))

# Filtrar datos
df_filtrado = df[(df['TIPO'] == tipo_seleccionado) & (df['QVENTAS'].between(rango_min, rango_max))]

# Recalcular deciles
if not df_filtrado.empty:
    df_resultado = asignar_deciles_y_rangos(df_filtrado)
    st.write(f"Datos filtrados y deciles recalculados para el tipo **{tipo_seleccionado}**:")
    st.dataframe(df_resultado)
else:
    st.write("No hay datos disponibles con el filtro seleccionado.")
```
